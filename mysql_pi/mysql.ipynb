```python

class simu_sql_link:
    def __init__(self, host='192.168.70.89', port=3306, user='root', password='jtwx1234', db='wz'):
        self.host = host
        self.port = port
        self.user = user
        self.password = password
        self.db = db
        self.conn = None
        self.cursor = None

    def start_link(self):
        self.conn = pymysql.connect(host=self.host, port=self.port, user=self.user, passwd=self.password, db=self.db)
        self.cursor = self.conn.cursor()

    def close_link(self):
        # 关闭游标
        self.cursor.close()
        # 关闭连接
        self.conn.close()

    def select_all(self, sheetname, iscolumn=False):
        """获取全部数据"""
        if not self.istableexist(sheetname):
            return None
        else:
            strcom = 'select * from ' + sheetname
            self.cursor.execute(strcom)
            result_tmp = self.cursor.fetchall()
            # print(result_tmp)
            if not result_tmp or len(result_tmp) == 0:
                return None
            length = len(result_tmp[0])
            result = []
            for i in range(len(result_tmp)):
                result_tmp2 = []
                for j in range(length):
                    result_tmp2.append(result_tmp[i][j])
                result.append(result_tmp2)

            if iscolumn:
                result.insert(0, self.get_column_name(sheetname))
            return result

    def select_last_but_one(self, sheetname, iscolumn=False):
        """获取倒数第二行"""
        if not self.istableexist(sheetname):
            return None
        else:
            strcommand = 'SELECT max(id) FROM ' + sheetname
            self.cursor.execute(strcommand)
            num = self.cursor.fetchone()[0]
            if not num:
                return None
            elif num > 2:
                strcommand = 'select * from ' + sheetname + ' where id = ' + str(num-1)
            else:
                strcommand = 'select * from ' + sheetname
            self.cursor.execute(strcommand)
            result_tmp = self.cursor.fetchone()
            result = []
            for i in range(len(result_tmp)):
                result.append(result_tmp[i])

            if iscolumn:
                result = [result]
                result.insert(0, self.get_column_name(sheetname))
            return result

    def get_column_name(self, sheetname):
        """获取列名"""
        if not self.istableexist(sheetname):
            return None
        else:
            strcommand = 'select COLUMN_NAME from information_schema.COLUMNS where table_name = \'' + sheetname + '\' and TABLE_SCHEMA=\'' + self.db + '\''
            self.cursor.execute(strcommand)
            result_tmp = self.cursor.fetchall()
            result = []
            for i in range(len(result_tmp)):
                result.append(result_tmp[i][0])
            return result

    def istableexist(self, sheetname):
        """"表名是否存在"""
        strcommand = 'SELECT table_name FROM information_schema.TABLES WHERE table_name = \'' + sheetname + '\' and TABLE_SCHEMA=\'' + self.db + '\''
        self.cursor.execute(strcommand)
        result_tmp = self.cursor.fetchall()
        if len(result_tmp) > 0:
            return True
        else:
            return False
```
