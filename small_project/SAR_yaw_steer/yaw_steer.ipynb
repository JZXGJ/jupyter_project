# 偏航导引分析
偏航角度与轨道参数有关

> Raney, R. K. "Doppler properties of radars in circular orbits." International Journal of Remote Sensing 7.9 (1986): 1153-1162.

$$
\begin{align}
\tan a_0 &= \frac{\omega / \omega_e - \cos \psi}{\cos \beta \sin\psi} \\
a_0 &= \arctan(\frac{\omega / \omega_e - \cos \psi}{\cos \beta \sin\psi})
\end{align}
$$
$\omega$: spacecraft orbital rotation rate

$\omega_e$: earth rotation rate

$\psi$: orbit inclination

$\beta$: argument of latitude

```python
import numpy as np
import astro
#print(dir(np))

psi     = np.deg2rad(97.42)
omega   = astro.orbit_angular_rate(6378e3+500e3)
omega_e = astro.earth_rotation_rate

def yaw_steer(beta):
    return np.arctan( (omega/omega_e - np.cos(psi)) / (np.cos(beta) * np.sin(psi)) )

beta = np.linspace(-np.pi/2, np.pi/2, 100)
yaw = yaw_steer(beta)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(np.rad2deg(beta), np.rad2deg(yaw))
ax.set_xlabel("argument of latitude (deg)")
ax.set_ylabel("yaw angle (deg)")
ax.grid()
fig.show()
fig.savefig("result.jpg", dpi=150)
```
