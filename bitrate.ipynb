```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

length = 100
n = np.zeros(length)
bitrate = np.zeros(length)
error = np.zeros(length-1)
bitrate_x = np.zeros(length-1)

for i in range(length):
    n[i] = i+1
    bitrate[i] = 59904000 / n[i]
    if i > 0:
        bitrate_x[i-1] = bitrate[i]
        error[i-1] = (n[i] - n[i-1]) / n[i]

plt.plot(bitrate_x, error*100, '-o')
plt.xlabel('bit rate (bps)')
plt.ylabel('bit rate error (%)')
plt.title('')
plt.grid(True)
plt.show()
```
